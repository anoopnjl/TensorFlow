```
import pandas as pd
import numpy as np
import tensorflow as tf
```

```
num_epoch = 600
mnist = tf.keras.datasets.mnist
(x_train,y_train),(x_test,y_test) = mnist.load_data()
#apply on_hot function on the Y variable for using softmax cross entropy
y_train_one_hot = tf.one_hot(y_train,10)
y_test_one_hot = tf.one_hot(y_test,10)
#normalize feature array to 2 dimension
x_train_norm = x_train.reshape(x_train.shape[0],784)
x_test_norm = x_test.reshape(x_test.shape[0],784)
```

```
#create MLP model with two hidden layers and output layer

#define the variables

X = tf.placeholder(tf.float32, [None,784])
y = tf.placeholder(tf.float32, [None, 10])

#define y_true_cls to hold y value to compare it against calculated values
y_true_cls = tf.placeholder(tf.int64, [None])

#first hidden layer
#W1 = tf.Variable(tf.random_normal([784,300], stddev=0.1))
#b1 = tf.Variable(tf.zeros([300]))
W1 = tf.Variable(tf.random_normal([784,300], stddev=0.1))
b1 = tf.Variable(tf.zeros([300]))
A1 = tf.nn.relu(tf.matmul(X,W1)+ b1)

#second hidden layer
W2 = tf.Variable(tf.random_normal([300,100], stddev=0.1))
b2 = tf.Variable(tf.zeros([100]))
A2 = tf.nn.relu(tf.matmul(A1,W2)+ b2)

#output layer
W3 = tf.Variable(tf.random_normal([100,10], stddev=0.1))
b3 = tf.Variable(tf.zeros([10]))
logits = tf.matmul(A2,W3)+ b3
#logits = tf.matmul(A1,W2)+ b2
y_pred = tf.nn.softmax(logits)   
y_pred_cls = tf.argmax(logits, axis=1)
```

```
#cross entropy for multiclassification
cross_entropy = tf.nn.softmax_cross_entropy_with_logits_v2(logits=logits, labels=y)
cost = tf.reduce_mean(cross_entropy)
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.0001).minimize(cost)
```

```
#calculating correct predictions and accuracy
correct_prediction = tf.equal(y_pred_cls, y_true_cls)
accuracy = tf.reduce_mean(tf.cast(correct_prediction,tf.float32))
```

```
#initialize session
session = tf.Session()
session.run(tf.global_variables_initializer())
#convert tesnsorflow object to python list for use in feed_dict
y_train_one_hot = session.run(y_train_one_hot)
y_test_one_hot = session.run(y_test_one_hot)
```

```
def optimize(num_iterations):
  for i in range(num_iterations):
    #for epoch in range(num_epoch):
      #batch_x = x_train_norm[epoch*100:(epoch+1)*100,:]
      #batch_y = y_train_one_hot[epoch*100:(epoch+1)*100,:]
      #feed_dict = {X:batch_x, y:batch_y}
      #session.run(optimizer, feed_dict=feed_dict)
    feed_dict = {X:x_train_norm, y:y_train_one_hot}
    session.run(optimizer, feed_dict=feed_dict)
    acc=session.run(accuracy, feed_dict={X:x_test_norm, y:y_test_one_hot, y_true_cls:y_test})
    print("Accuracy on test-set after iteration {0:d}: {1:.1%}".format(i,acc))
```

```
optimize(250)
session.close
```
